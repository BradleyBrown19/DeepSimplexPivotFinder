```python
from itertools import permutations
 
# TSP Solver (brute force)
def TSP(G, N):
    min_weight = float("inf")
    min_path = None

    for perm in permutations(list(range(1,N))):
      current_pathweight = G[0][perm[0]]
      for i in range(len(perm)-1):
          current_pathweight += G[perm[i]][perm[i+1]]
      current_pathweight += G[perm[-1]][0]
      # update
      if current_pathweight < min_weight:
        min_weight = current_pathweight
        min_path = perm

    return min_weight, [0] + list(min_path)
```

```python
# THIS CELL CONTAINS UTIL GRAPH FUNCTIONS

# print adjacency matrix (distances rounded to 2 decimal)
def print_2d(M):
  for r in M:
    print([round(x,2) for x in r])
  print()

# generate N 2D points
# then convert all points to (v1, v2, w(v1, v2)) tuples to represent graph
def generate_graph(N, D):
  def dist(a, b):
      return math.sqrt( (a[0]-b[0])**2 + (a[1]-b[1])**2 )
  # generate 2d points
  points = set()
  while len(points) < N:
    points.add((np.random.uniform(0,D), np.random.uniform(0,D)))
  points = np.array(list(points))
  # generate graph
  G = set()
  W = {}
  for i in range(N):
    for j in range(i+1, N):
      d = dist(points[i], points[j])
      G.add((i, j, d))
      W[i,j] = d
      W[j,i] = d
  return points, G, W

# visualize 2D points
def visualize_points(points, D):
  plt.scatter([p[0] for p in points[1:]], 
              [p[1] for p in points[1:]], c="blue")
  plt.scatter(points[0][0], points[0][1], c='red')
  plt.ylim([0, D])
  plt.xlim([0, D])
  plt.show()

# convert graph represented in tuples to graph represented in
# adjacency matrix
def tuples2adj(G, N):
  M = [[0]*N for _ in range(N)]
  for u,v,w in G:
    M[u][v] = w
    M[v][u] = w
  return M
```

```python
# STEP 1: DEFINE PARAMS
N = 8 # number of nodes in TSP
D = 10 # all 2D points lie within x=(0,D) and y=(0,D)
# the generated graph will be a set of tuples (v1, v2, w(v1,v2))

# STEP 2: GENERATE GRAPH
points, G, W = generate_graph(N, D)
print("Generated Locations:")
visualize_points(points, D)
M = tuples2adj(G, N)
print("\nAdjacency Matrix:")
print_2d(M)

# STEP 3: SOLVE TSP
# path = TSP(np.array(points), 0.001) # vertex ordering
dist, path = TSP(M, N)
path.append(path[0])

# ordered_point = np.concatenate((points[path], [points[0]])) # point ordering
ordered_point = points[path]
print("path: " + str(path))
print("distance: " + str(dist))

# STEP 4: VISUALIZE
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 1, 1])
axes.set_xlim(0, D)
axes.set_ylim(0, D)
axes.plot(points[0][0], points[0][1], marker='o', c='red') # starting point
# plot rest of the points
for i in range(1,N):
  axes.plot(points[i][0], points[i][1], marker='o', c="blue")
# draw used edges
axes.plot(ordered_point[:,0], ordered_point[:,1])
fig.show()
```

```python
from scipy.optimize import linprog

# fill in c
c = []
for i in range(N):
  for j in range(N):
    if i != j:
      c.append(W[i,j])
c += [0]*(N-1)

# fill in A_eq
A_eq = [[0]*((N+1)*(N-1)) for _ in range(2*N)]
k = 0
for i in range(N):
  for j in range(N):
    if i != j:
      A_eq[i][k] = 1
      A_eq[j+N][k] = 1
      k += 1

# fill in A_ub
A_ub = [[0]*((N+1)*(N-1)) for _ in range((N-1)*(N-2))]
k = 0
q = 0
for i in range(N):
  for j in range(N):
    if i == j:
      continue
    if i == 0 or j == 0:
      k += 1
      continue
    A_ub[q][k] = N
    A_ub[q][(N*(N-1))+i-1] = 1
    A_ub[q][(N*(N-1))+j-1] = -1
    k += 1
    q += 1

# fill in b_eq
b_eq = [1]*(2*N)

# fill in b_ub
b_ub = [N-1]*((N-1)*(N-2))

# print("c:")
# print(c, "\n")
# print("A_eq:")
# print_2d(A_eq)
# print("A_ub:")
# print_2d(A_ub)
# print("b_eq:")
# print(b_eq, "\n")
# print("b_ub:")
# print(b_ub, "\n")

# simplex
# res = linprog(c=c, 
#               A_eq=A_eq, 
#               b_eq=b_eq, 
#               A_ub=A_ub, 
#               b_ub=b_ub, 
#               method="simplex")
# assert(res["message"] == 'Optimization terminated successfully.')
# assert(res["success"] == True)
# print(res["x"]) # edge selection (fractional)
# print(res["fun"]) # objective function value

soln = float("inf")
best_x = None

stack = [(A_eq, b_eq)]
while stack:
  A_eq, b_eq = stack.pop()
  res = linprog(c=c, A_eq=A_eq, b_eq=b_eq, A_ub=A_ub, b_ub=b_ub, method="simplex")
  x = res["x"]
  fun = res["fun"]
  status = res["status"]
  if status != 0:
    continue

  # check which Xij is not 0 or 1
  i = 0
  while i < N*(N-1):
    if abs(x[i]) > 0.001 and abs(x[i]-1) > 0.001:
      break
    i += 1

  if i == N*(N-1) and fun < soln:
    soln = fun
    best_x = x
  elif fun < soln:
    r = [0]*((N+1)*(N-1))
    r[i] = 1
    stack.append((A_eq+[r], b_eq+[1]))
    stack.append((A_eq+[r], b_eq+[0]))

print(best_x)
print(soln)
```
